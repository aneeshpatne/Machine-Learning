```python
import pandas as pd
df = pd.read_csv('potato_data.csv')
df.head()
```

```python
df['Price'].isna().any()
```

# dropping NaN values

```python

df_new = df.dropna()
df_new
```

# Replace NaN values

```python

#replace with 0 
df_new = df.fillna(value=0)
df_new
```

```python
#mean
df_new = df.fillna(df['Price'].mean())
df_new
```

```python
#median
df_new = df.fillna(df['Price'].median())
df_new
```
